```python
import pandas as pd
import numpy as np
import os
import pathlib
import shutil
```

```python
# Config
LABELS = 'labels.csv'
dogs_dir = 'dogs/'
```

```python
labels = pd.read_csv(LABELS)
```

```python
# see the labels
labels.head()
```

```python
labels = dict(zip(list(labels.id), list(labels.breed)))
```

```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
```

```python
#Create a root folder for all the dog breeds
createFolder('./dog_photos')
```

```python
for item in os.listdir(dogs_dir):
    if not item.startswith('.'):
        print('preprocessing: '+item)
        #Get the dog_id from the filename
        dog_id = item.split('.')[0]
        
        #Get the breed name from the dictionary using the dog_id from the previous step
        breed = labels[dog_id]
        
        #Create a folder
        destination = './dog_photos/'+breed
        createFolder(destination)
        
        #Copy the dog image into the new folder
        shutil.copy(dogs_dir+item, destination)          
```
